### 3) target 데이터



```python
# target 분포
df.point.value_counts()
```

```python
# # 1,2 점은 0(부정), 3,4 점은 1(긍정)
li = [1,2,4,5]
change = [0,0,1,1]
df['point']=df['point'].replace(li,change)

#replace(바꿀패탄, 덮어쓸 패턴)
# li = [1,2,4,5]
# change = [ 0,0,1,1]
# => 여러번 처리해야하는걸 한번에 리스트로 묶어서 처리하는걸 fancy indexing
```

### 4) 불용어 처리 및 토큰화

```python
# 한글과 공백을 제외하고 모두 제거
df['review']=df['review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") # [^ㄱ-ㅎㅏ-ㅣ가-힣 ] 마지막에 띄어쓰기 해주
```

```python
# null값 다시 생겼는지 확인
df['review'] = df['review'].str.replace('^ +', "") # white space 데이터를 empty value로 변경
df['review'].replace('', np.nan, inplace=True)
print(df.isnull().sum())
```

```python
# 불용어 사전 만들기
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
!pip install konlpy
from konlpy.tag import Okt
```

```python
okt=Okt()
tok_data=[]
for sent in df['review']:
    tok_sent=okt.morphs(sent) #토큰화
    sw_rem_sent=[w for w in tok_sent if not w in stopwords]
    tow_rem_sent = [w for w in sw_rem_sent if len(w) >= 2 ]
    tok_data.append(tow_rem_sent)
```

```python
df['tok_word'] = tok_data
```

```python
df
```

```python
tok_data
```

```python
# 데이터 셋 나누기

xdata= df['tok_word']
ydata = df['point']
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(xdata,ydata,
                                              test_size=0.2,
                                              random_state=1121,
                                              stratify = ydata)
```

```python
xtrain.shape # 159925
xtest.shape # 39982
type(xtrain)
```

```python
tok = Tokenizer()
tok.fit_on_texts(xtrain) # xtrain을 코퍼스로하여 fitting(단어별로 인덱스가 부여됨)
xtrain_enc = tok.texts_to_sequences(xtrain)
```

```python
xtrain_enc
```

```python
xtrain_enc_idx = tok.word_index
xtrain_enc_idx
```

```python
len(xtrain_enc_idx) # 95617 단어로 구성
```

```python
# 단어 빈도수가 2이하인 것은 제거
# 0번 패딩 토큰을 고려하여 + 1

vocab_size = len(xtrain_enc_idx)+1
vocab_size # 95618
```

### 5) 패딩

```python
# 단어 최대길이 찾기
maxLen = max(map(len, xtrain_enc))
maxLen
# 방법 2 # max(map(len, xtrain_enc)) for 문
# maxLen = max(len(s) for s in xtrain_enc)
```

```python
xtrain_padded=pad_sequences(xtrain_enc, maxlen=maxLen)
xtrain_padded.shape # (159925, 47)
xtrain_padded[0]
```

## 3. 모델 만들기

```python
from tensorflow.keras.layers import SimpleRNN, Embedding, Dense,LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 100))
model.add(LSTM(128))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
# val_loss 가 중가 -> 과적합 가능성 큼 , 데이터 손실이 4회 이상이면 종료
mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
# val_acc 가 이전보다 좋아질 경우 모델 저장
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(xtrain_padded, ytrain, epochs=10, callbacks=[es, mc], batch_size=64, validation_split=0.2)
```

```python
xtest_enc = tok.texts_to_sequences(xtest)
xtest_padded = pad_sequences(xtest_enc,maxlen = maxLen)
```

```python
loaded_model = load_model('best_model.h5')
loaded_model.evaluate(xtest_padded, ytest)[1]
```
